# Insert Hawker Lat-Long Data

This sheet aims to collect all hawker centre data in Singapore, together with their long and lat for data processing

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

import re
```

```python
import requests
import json
import keys

url = "https://developers.onemap.sg/privateapi/themesvc/retrieveTheme?queryName=hawkercentre&token="+keys.api_key

resp = requests.get(url)

#Convert JSON into Python Object 
data=json.loads(resp.content)
data
```

```python
data['SrchResults']
```

```python
df_hawkers = pd.DataFrame(columns=['Name', 'Lat-Long'])
```

```python
for i in range(1, len(data['SrchResults'])):
    df_hawkers = df_hawkers.append({'Name':data['SrchResults'][i]['NAME'], 'Lat-Long':data['SrchResults'][i]['LatLng']}, ignore_index=True)
```

```python
df_hawkers[["Latitude", "Longitude"]] = df_hawkers["Lat-Long"].str.split(",", expand=True)
```

```python
df_hawkers = df_hawkers.drop(columns=['Lat-Long'])
```

```python
df_hawkers.to_csv('hawkers_data.csv', index=False)
```
